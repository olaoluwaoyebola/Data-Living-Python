```python
fruits = ['banana', 'grape', 'pear', 'watermelon', 'mango', 'pineapple']
idxs = [5, 4, 2]
popped = [fruits.pop(i) for i in sorted(idxs, reverse=True)]
print(popped)
```

```python
fruits = ['banana', 'grape', 'pear', 'watermelon', 'mango', 'pear', 'pineapple']
popped = fruits[2:5]
del fruits[2:5]
print(popped)
print(fruits)
```

```python
fruits = ['banana', 'grape', 'pear', 'watermelon', 'mango', 'pear', 'pineapple']
for name in ("mango", "pear", "pineapple"):
    while name in fruits:
        fruits.remove(name)
print(fruits)
```
