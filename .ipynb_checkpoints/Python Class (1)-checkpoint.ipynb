```python
guesses = 0
secret_number = 7
guess_limit_number = 3
while guesses < guess_limit_number:
    #print("Guess my number! Hint: 7")
    guess = int(input())
    if guess == secret_number:
        print("Yes!")
        break
    guesses += 1 # Add 1
else:
    print("Too many tries!")
```

# Error Handling

```python
a = 10 / 0
print(a)
```

```python
def safe_divide(a, b):
    try:
        result = a / b
    except ZeroDivisionError: # specific error
        print("Oops! Can't divide by zero - try again!")
        result = "Error"
    else:
        print("Yay, it worked!")
    finally:
        print("Math time over -  back to play!")
    return result

print(safe_divide(10, 2)) # 5.0 - Yay!
# print(safe_divide(10, 0)) 
```

```python
def safe_divide(a, b):
    try:
        result = a / b
    except ZeroDivisionError: # specific error
        print("Oops! Can't divide by zero - try again!")
        result = "Error"
    else:
        print("Yay, it worked!")
    finally:
        print("Math time over -  back to play!")
    return result

# print(safe_divide(10, 2)) # 5.0 - Yay!
print(safe_divide(10, 0)) 
```

```python
def safe_divide(a, b):
    try:
        result = a / b
    except Exception as e:
        print("Oops! Can't divide by zero - try again!")
        result = "Error"
    else:
        print("Yay, it worked!")
    finally:
        print("Math time over -  back to play!")
    return result

# print(safe_divide(10, 2)) # 5.0 - Yay!
print(safe_divide(10, 0)) 
```
