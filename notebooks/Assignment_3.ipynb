```python
# simple filtering

df_prices = pd.DataFrame({
    'Toy': ['Ball', 'Car', 'Doll', 'Train'],
    'Price': [10.0, 20.0, 30.0, 45.0]
})

df_stock = pd.DataFrame({
    'Toy': ['Ball', 'Car', 'Doll', 'Robot'],
    'Stock': [50, 40, 30, 10]
})
```

### Task 4

```python
pd.merge(df_prices, df_stock, how = "inner")
```

```python
# outer merge

pd.merge(df_prices, df_stock, how = "outer")
```
